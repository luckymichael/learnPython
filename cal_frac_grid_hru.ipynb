```python
import pandas as pd
import numpy as np
import geopandas as gp
import os
os.chdir(r'D:\Cloud\Dropbox\postdoc\summa\summaData\blivneh')
```

```python
# read fraction table of covered hrus
df_frac = gp.read_file('grid_hru_intersec.dbf')
df_frac = df_frac.drop(['geometry', 'lat', 'lon'], 1)
df_frac.columns = ['Area', 'hruID' , 'gridID']
df_frac = df_frac[['hruID' , 'gridID', 'Area']]
```

```python
# read uncovered hrus
df_lone = pd.read_csv('hru_not_covered.csv')
df_lone.columns = ['hruID' , 'gridID', 'Area']
df_lone.Area = 1000.0
```

```python
# read grid cells
df_grid = gp.read_file('blivneh_grid_crb_proj.dbf')
df_grid.index = df_grid.id
df_grid['x'] = (df_grid.xmin + df_grid.xmax) * 0.5
df_grid['y'] = (df_grid.ymin + df_grid.ymax) * 0.5
```

```python
# concatenate two tables
df_all = df_frac.append(df_lone)
df_all.hruID = df_all.hruID.astype(int)
df_all.gridID = df_all.gridID.astype(int)
```

```python
# calculate the lat lon for each fraction
df_all['lat'] = df_grid.y[df_all.gridID].values
df_all['lon'] = df_grid.x[df_all.gridID].values
```

```python
group_hru = df_all.groupby('hruID')
hru_area = group_hru.Area.sum()
df_all['frac'] = df_all.Area / hru_area[df_all.hruID].values
```
